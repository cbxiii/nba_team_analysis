```python
import pandas as pd
from nba_api.stats.endpoints import teamgamelog
```

```python
heat_gamelog = teamgamelog.TeamGameLog(team_id='1610612748', season='2022')
```

```python
df = heat_gamelog.get_data_frames()[0]
df.head(10)
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.head(10)
```

```python
df = df.drop(columns='MIN')
df
```

```python
split_df = df['MATCHUP'].str.split(' ', expand=True)
split_df
```

```python
df['HOME_AWAY'] = split_df[1].apply(lambda x: 'HOME' if x == 'vs.' else 'AWAY')
df
```

```python
df['OPP'] = split_df[2]
df
```

```python
from matplotlib import pyplot as plt
plt.plot(df['GAME_DATE'], df['W_PCT'])
plt.xlabel('Time')
plt.ylabel('Win %')
plt.legend()
plt.show()
```

```python
chron_df = df.reindex(index=df.index[::-1])
chron_df
```

```python
df = chron_df
plt.plot(df['GAME_DATE'], df['W_PCT'])
plt.xlabel('Game Date')
plt.ylabel('Win Percentage')
plt.title('Win Percentage Over the Course of the Season')
plt.show()
```

```python
plt.hist(x=df['PTS'], bins=7, color='red', edgecolor='black')
plt.xlabel('Points Scored')
plt.ylabel('Frequency')
plt.title('Points Scored Distribution')
```
